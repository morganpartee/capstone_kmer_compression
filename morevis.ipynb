```python
import pandas as pd
import seaborn as sns
from datetime import datetime
```

```python
results = pd.read_csv("capstone_kmer_compression/0.4.csv", header=None, names = ["Token Size", "K-Size", "Initial Features", "Resultant Features", "Accuracy", "SD", "Detailed_Results","Start Time", "End Time"])
results['Start Time'] = results['Start Time'].apply(lambda x: datetime.fromisoformat(x[1:]))
results['End Time'] = results['End Time'].apply(lambda x: datetime.fromisoformat(x[1:]))
results['Duration'] = (results['End Time'] - results['Start Time']).apply(lambda x: x.seconds/60)
```

```python
results.head()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x='K-Size', y='Accuracy', hue="Token Size", data=results, palette="deep")
```

```python
sns.lineplot(x='K-Size', y='Duration', hue="Token Size", data=results, palette="deep")
```
